```python
import pandas as pd 
import numpy as np

aos_light = pd.read_csv('raw_aos_data/aos_ANDROID_LIGHT.csv', index_col = 0)
aos_ema = pd.read_csv('raw_aos_data/aos_ema.csv', index_col = 0)
```

```python
aos_sensor_removed = ['INSTAGRAM_FEATURES', 'MUSIC_PLAYING', 'UNKNOWN', 'LOCATIONS_MANUAL', 'ANDROID_HEART_RATE', 'SLEEP']
aos_sensors = list(aos_dict.keys() - set(aos_sensor_removed))

ios_sensor_removed = ['APPLICATION_USAGE', 'ANDROID_WIFI', 'INSTAGRAM_FEATURES', 'KEYSTROKE_LOG', 'NOTIFICATIONS', 'SMS', 'TYPING', 'ANDROID_LIGHT', 'MUSIC_PLAYING', 'UNKNOWN', 'LOCATIONS_MANUAL', 'ANDROID_HEART_RATE', 'SLEEP']
ios_sensors = list(ios_dict.keys() - set(ios_sensor_removed))

print('Android sensors:', aos_sensors)
print('iOS sensors:', ios_sensors)

for s in aos_sensors:
    aos_sensor_list = list()
    for df in aos_dict[s]:
        aos_sensor_list.append(df)
    aos_df = pd.concat(aos_sensor_list)
    
    aos_df['timestamp'] = aos_df.index
    aos_df.index = pd.to_datetime(aos_df.timestamp, unit='ms', utc=True).dt.tz_convert('Asia/Seoul')
    aos_df.drop('timestamp', inplace=True, axis=1)
```

```python
aos_light['timestamp'] = aos_light.index
aos_light.index = pd.to_datetime(aos_light.timestamp, unit='ms', utc=True).dt.tz_convert('Asia/Seoul')
aos_light.drop('timestamp', inplace=True, axis=1)

aos_ema['timestamp'] = aos_ema.index
aos_ema.index = pd.to_datetime(aos_ema.timestamp, unit='ms', utc=True).dt.tz_convert('Asia/Seoul')
aos_ema.drop('timestamp', inplace=True, axis=1)
```

```python
import plotly.graph_objs as go

aos_pids = list(set(aos_ema['pid']))
fig = go.Figure()

i = 0
for pid in aos_pids[:2]:
    i += 1
    fig.add_trace(go.Scatter(x=aos_light.loc[aos_light.pid == pid].index, y=np.ones(aos_light.loc[aos_light.pid == pid].shape[0]), mode = 'markers', name= "light_p_{}".format(pid), yaxis="y{}".format(i)))
    fig.add_trace(go.Scatter(x=aos_ema.loc[aos_ema.pid == pid].index, y=np.ones(aos_ema.loc[aos_ema.pid == pid].shape[0]), mode = 'markers', name= "ema_p_{}".format(pid), yaxis="y{}".format(i)))
  
    fig.update_layout(
        xaxis=dict(
            autorange=True, 
            rangeslider=dict(
                autorange=True, 
                ),
                type="date"
            ),

            yaxis1=dict(
                anchor="x",
                autorange=True,
                domain=[0.3, 0.6], 
                mirror=True, 
                showline=True,
                side="right",
                showticklabels =False,

                tickmode="auto",
                zeroline=False
            ),

            yaxis2=dict(
                anchor="x",
                autorange=True,
                domain=[0.6, 0.9], 
                mirror=True, 
                showline=True,
                side="right",
                showticklabels =False,

                tickmode="auto",
                zeroline=False
            ))

fig.update_xaxes(
    rangeslider_visible=True, 
    rangeselector=dict(
        buttons=list([
                dict(count=1, label="s", step="second", stepmode="backward"), 
                dict(count=15, label="15s", step="second", stepmode="backward"),
                dict(count=30, label="30s", step="second", stepmode="backward"),
                dict(count=1, label="1m", step="minute", stepmode="backward"),
                dict(step="all")
            ])
        )
    )

fig.update_layout(
        dragmode="zoom",
        hovermode="x",
        legend=dict(traceorder="reversed"),
        height=500,
        template="plotly_white",
        margin=dict(
            t=100,
            b=100
        ),
    )

fig.show()
```

```python
import plotly.graph_objs as go



aos_pids = list(set(aos_ema['pid']))

for s in aos_sensors[:1]:
    aos_sensor_list = list()
    for df in aos_dict[s]:
        aos_sensor_list.append(df)
    aos_df = pd.concat(aos_sensor_list)
    
    aos_df['timestamp'] = aos_df.index
    aos_df.index = pd.to_datetime(aos_df.timestamp, unit='ms', utc=True).dt.tz_convert('Asia/Seoul')
    aos_df.drop('timestamp', inplace=True, axis=1)

    
    for pid in aos_pids[:3]:
        fig = go.Figure()
        # Add traces
        fig.add_trace(go.Scatter(x=aos_df.loc[aos_df.pid == pid].index, y=np.ones(aos_df.loc[aos_df.pid == pid].shape[0]), mode = 'markers', name= "{}_p_{}".format(s, pid), yaxis="y1"))
        fig.add_trace(go.Scatter(x=aos_ema.loc[aos_ema.pid == pid].index, y=np.ones(aos_ema.loc[aos_ema.pid == pid].shape[0]), mode = 'markers', name= "ema_p_{}".format(pid), yaxis="y1"))

        # Update axes
        fig.update_layout(
            xaxis=dict(
                autorange=True, # Determines whether or not range of axis is computed in relation to input data. one of ( True | False | "reversed" )
                rangeslider=dict(
                    autorange=True, 
                ),
                type="date" # Sets the axis type, reference: https://plotly.com/python/reference/layout/xaxis/#layout-xaxis-type
                            # one of ( "-" | "linear" | "log" | "date" | "category" | "multicategory" )
            ),

            yaxis1=dict(
                anchor="x",
                autorange=True,
                domain=[0.3, 0.7], # Set visualization range for sub plot
                mirror=True, # Determines if the axis lines or/and ticks are mirrored to the opposite side of the plotting area. 
                            # one of ( True | "ticks" | False | "all" | "allticks" )
                            # reference: https://plotly.com/python/reference/layout/xaxis/#layout-xaxis-mirror
                showline=True,
                side="right",
                showticklabels =False,

                tickmode="auto", # Sets the tick mode for this axis, one of ( "auto" | "linear" | "array" ), 
                                # reference: https://plotly.com/python/reference/layout/xaxis/#layout-xaxis-tickmode
                zeroline=False
            ))

        fig.update_xaxes(
            rangeslider_visible=True, 
            rangeselector=dict(
                buttons=list([
                    dict(count=1, label="s", step="second", stepmode="backward"), 
                    dict(count=15, label="15s", step="second", stepmode="backward"),
                    dict(count=30, label="30s", step="second", stepmode="backward"),
                    dict(count=1, label="1m", step="minute", stepmode="backward"),
                    dict(step="all")
                ])
            )
        )

        # Update layout
        fig.update_layout(
            dragmode="zoom",
            hovermode="x",
            legend=dict(traceorder="reversed"),
            height=500,
            template="plotly_white",
            margin=dict(
                t=100,
                b=100
            ),
        )

        fig.show()
```
